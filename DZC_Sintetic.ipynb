```python
#Librerias a importar
import pandas as pd # Manipular conjunto de datos llamados dataframe
import numpy as np # Operaciones con matrices
import matplotlib.pyplot as plt # Gráficas
```

```python
#se define un data frame para cargar el archivo que contiene los datos de entrenamiento
datos1 = pd.read_csv("/UCordoba/Machine Learning/Data/datasetDZC_Sintetic_.csv")
#se muestra el dataframe
datos1
```

```python
datos1.info()
```

```python
#Imputación de variables numéricas:media
from sklearn.impute import SimpleImputer
ImpNumeros = SimpleImputer(missing_values=np.nan, strategy='mean') #reemplaza por la media
columnas_a_imputar = ['Dias_sintomas', 'edad']
datos1[columnas_a_imputar] = ImpNumeros.fit_transform(datos1[columnas_a_imputar])
datos1.info()
```

```python
#Imputacion de variables categóricas: moda
from sklearn.impute import SimpleImputer
ImpCategorias = SimpleImputer(missing_values=np.nan, strategy='most_frequent') #se reemplaza por la moda
datos1[['CS_RACA',"GESTANT"]] = ImpCategorias.fit_transform(datos1[['CS_RACA',"GESTANT"]])
datos1.info()
```
