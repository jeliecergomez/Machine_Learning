```python
#Se crea una estructura de un dataframe para cargar los datos
import pandas as pd   
#Esta librería permite la estimación de modelos estádisticos
import statsmodels.api as sm   
#Esta librería ajusta los modelos que se utilizan basados en formulas de estilo R
import statsmodels.formula.api as smf 
# Permite cargar el modelo de regresión lineal 
from sklearn import linear_model
```

```python
#se define un data frame para cargar el archivo que contiene los datos de entrenamiento
apartamentos = pd.read_csv("/Machine Learning/Data/apartamentos_los_angeles.csv")
#se muestra el dataframe
apartamentos.head(8)
```

```python
%matplotlib inline
#Esta librería permite la generación de gráficos 
from matplotlib import pyplot              
#Esta librería permite generar gráficos en 3D
from mpl_toolkits.mplot3d import Axes3D   


myfigure = pyplot.figure(figsize=(7, 5))       
ax = Axes3D(myfigure)
#Datos del eje X
x1 = apartamentos["area"]   
#Datos del eje Y
x2 = apartamentos["habitaciones"]   
#Datos del eje Z (precio estimado)
y = apartamentos["precio"]                              

ax.scatter(x1, x2, y, marker='*', c='r')
 # Etiqueta para el eje X
ax.set_xlabel('area en metros cuadrados') 
# Etiqueta para el eje Y
ax.set_ylabel('# de habitaciones') 
 # Etiqueta para el eje Z
ax.set_zlabel('precio');    
```

```python
# Ajuste del modelo 
modelo_ajustado = smf.ols('precio ~ area + habitaciones + edad', data=apartamentos).fit()  
print(modelo_ajustado.summary())
```

```python
modelo_ajustado.params
```

```python
#Se define el modelo de regresión líneal
reg=linear_model.LinearRegression()
#Se solicita el ajuste del modelo de regresión líneal basados en los parámetros de
#area, habitaciones y edad, para calcular el precio
reg.fit(apartamentos[['area','habitaciones','edad']],apartamentos.precio)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[500,10,15]])
```
