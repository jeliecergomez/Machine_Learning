```python
import pandas as pd
#se define un data frame para cargar el archivo que contiene los datos de entrenamiento
dthealth = pd.read_csv("/Machine Learning/Data/Data Heart/heart.csv")
#se muestra el dataframe
dthealth.head()
```

```python
#Se almacena en la variable inputs la variables independientes y en mytarget la variable dependiente
inputs=dthealth.drop('target', axis='columns')
mytarget=dthealth['target']
```

```python
#Se visualizan las variables independientes
inputs
```

```python
#Cargamos el modelo con los datos del entrenamiento
from sklearn import tree
mymodeltree =tree.DecisionTreeClassifier()
mymodeltree.fit(inputs,mytarget)
```

```python
#Mostramos el puntaje de la predicción
mymodeltree.score(inputs,mytarget)
```

```python
#Se prueba el modelo con variables que permitan determinar la presencia de la enfermedad cardiaca 
mymodeltree.predict([[63,1,3,145,223,1,0,150,0,2.3,0,0,1]])
```

```python
#Se prueba el modelo con variables que permitan determinar la ausencia de la enfermedad cardiaca 
mymodeltree.predict([[56,1,2,130,256,1,0,142,1,0.6,1,1,1]])
```
